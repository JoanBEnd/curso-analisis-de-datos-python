# Creación de un Gráfico de Barras

Para comenzar, definiremos dos listas y las graficaremos:

- **`lista_año`**: Representará los años de ventas y será colocada en el eje **x**.
- **`lista_venta`**: Representará el total de ventas por año y será colocada en el eje **y**.

A continuación, usaremos la función **`plt.bar()`** para crear la gráfica de barras.  
Además, complementaremos la visualización con las siguientes funciones:

1. **`plt.figure(figsize=(10,5))`**: Nos permite aumentar el tamaño de la gráfica.  *Con los valores en el figsize, podemos jugar, para ver distintos tamaños
2. **`plt.xlabel()`**: Nos permite agregar un título al eje **x**, para identificar qué se está mostrando.
3. **`plt.ylabel()`**: Nos permite agregar un título al eje **y**, para identificar qué se está mostrando.
4. **`plt.title()`**: Nos permite agregar un título general al gráfico.

Estas funciones  nos ayudan a tener una visión más clara y legible para la gráfica.

```python
#Gráfico de barras

lista_año = [2020,2021,2022,2023,2024]
lista_venta = [450000,321000,523000,142000,652000]
plt.figure(figsize=(10,5))
plt.bar(lista_año ,lista_venta, color = 'blue',  alpha=0.5)
plt.xlabel('Años')  
plt.ylabel('Ventas')
plt.title('Ventas por año')
```

### Gráficas de Barras Horizontales

También podemos graficar barras horizontales utilizando la función **`plt.barh()`**.  

Además, podemos usar algunas funciones adicionales para mejorar la apariencia del gráfico:  

- Dentro de **`plt.barh()`**:
  - **`color`**: Nos permite definir el color de las barras.
  - **`alpha`**: Nos permite controlar la transparencia de las barras (valores entre 0 y 1).

- Dentro de **`plt.title()`**:
  - **`fontsize`**: Nos permite ajustar el tamaño del texto del título.



```python
#Gráfico de barras horizontal
plt.figure(figsize=(10,5))
plt.barh(lista_año ,lista_venta, color = 'red',  alpha=0.5)
plt.xlabel('Años')
plt.ylabel('Ventas')
plt.title('Ventas por año', fontsize=12)
```

### Gráficas con Datos desde un Archivo CSV

Hasta ahora hemos aprendido a graficar utilizando ejemplos con listas.  

En esta ocasión, trabajaremos con un archivo **CSV** para obtener la información.  
Usaremos la biblioteca **pandas** para tratar los datos y luego proceder a graficarlos.  
Con pandas sabemos que la información estará almacenada en un **DataFrame**, y que a partir de ello extraremos los datos para realizar la gráfica.


```python
mi_modalidad_frame = pd.read_csv("data/ModalidadVirtual.csv", parse_dates=["time"])
mi_modalidad_frame.head()
```

A continuación procedemos a calcular el total de carreras que han elegido por modalidad virtual.  

Con Pandas: 

1. Procedemos a calcular el total de valores, filtrando por el campo Carrera
2. Lo asignamos a una nuevo DataFrame
3. Renombramos los nombres de las columnas


Con respecto a la visualización:

1. Asignamos un tamaño a la gráfica.
2. Creamos la gráfica de barras, usando las 2 columnas del nuevo dataframe, de las carreras y su total.
3. Se asignan los titulos para cada eje
4. Se crea un titulo para la grafica y un tamaño de letra
5. la nueva funcion que usamos es:  
    **`plt.xticks(rotation=45, ha='right', fontsize=8)`**: Como veremos el texto por cada carrera son largas, va a ser imposible verlo bien, por eso usamos esta función, para darle una rotación al texto, ajustar la alineación y darle un tamaño al texto.


```python
#Pandas
mis_carreras_frame = mi_modalidad_frame["carrera"].value_counts().reset_index()
mis_carreras_frame.columns = ["carrera", "Total"]

#Gráfico de barras 
plt.figure(figsize=(20,8))
plt.bar(mis_carreras_frame["carrera"], mis_carreras_frame["Total"], color = 'green',  alpha=0.5)
plt.xlabel('Carreras')
plt.ylabel('Total')
plt.title('Carreras por modalidad virtual', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=8)

```

Ok. ya pudimos graficar los datos extraidos desde un csv.  
Ahora podremos graficar los datos con una barra horizontal.   
Será parecido a la gráfica que hizimos con las listas, solo que en este caso, reemplazaremos por los campos del dataframe.

```python
plt.figure(figsize=(20,10))
plt.barh(mis_carreras_frame["carrera"], mis_carreras_frame["Total"], color = 'purple',  alpha=0.5)
plt.xlabel('Carreras')
plt.ylabel('Total') 
plt.title('Carreras por modalidad virtual', fontsize=12)
```
