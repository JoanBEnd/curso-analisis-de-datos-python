### BOXPLOT

Un **boxplot** es una herramienta gráfica que nos permite analizar la distribución y asimetría de los datos. A continuación, explicamos sus componentes principales:

- **Asimetría en los datos:**  
  Si la mediana (línea naranja) está más cerca de Q1 o Q3, podemos identificar una asimetría en la distribución.

- **Distribución de los datos:**  
  - **Mediana:** Representada por la línea dentro de la caja, indica dónde se concentra la mayor cantidad de datos.  
  - **Cuartiles (Q1 y Q3):** Nos proporcionan información sobre cómo se distribuyen los datos dentro del rango intercuartílico (IQR).

- **Dispersión de los datos:**  
  - **Bigotes:** Muestran hasta dónde se extienden los datos, proporcionando una idea de su dispersión. Esto nos ayuda a identificar si los datos están muy dispersos o concentrados.

Un boxplot también facilita la detección de valores atípicos y posibles patrones en la distribución. Es una herramienta fundamental en el análisis exploratorio de datos.


```python


edades = [5, 8, 9, 10, 11, 20, 22, 24, 27, 30, 40, 50, 60, 66, 77, 80, 95]

#crear Boxplot
plt.figure(figsize=(12, 8))
plt.boxplot(edades)
plt.yticks(range(0, 101, 5))
plt.grid()
```

### Valores atípicos

Los valores atípicos también considerados **Outliers** son los que aparecen fuera del rango entre el Q1 y Q3, y se ven graficados como puntos en el boxplot. 

```python
edades = [3, 40, 44, 50, 52, 55, 60, 66, 98, 95]

#crear Boxplot

plt.figure(figsize=(12, 8))
plt.boxplot(edades)
plt.yticks(range(0, 101, 5))
plt.grid()

#los valores atipicos serian 3, 95 y 98
```

### Conceptos Básicos Representados en la Imagen

1. **Rango Intercuartílico (RIQ):**
    - Es la distancia entre el primer cuartil (Q1) y el tercer cuartil (Q3):  
      <div align="center" style="font-size: 24px;" > RIQ = Q3 - Q1</div>  
    - Este rango abarca el 50% central de los datos, es decir, donde se concentra la mayoría de los valores.

2. **Mediana:**
    - Es la línea roja dentro del boxplot (en el centro del rectángulo).
    - Representa el valor medio de los datos cuando están ordenados.
    - Nos indica dónde está la mitad de los datos.
    
<div align="center">
    <img src="imagen/mediana.png" alt="mediana" title="mediana" width="550">
</div>

3. **Bigotes (líneas que salen del rectángulo):**
    - Representan la dispersión de los datos, mostrando hasta dónde llegan los valores normales.
    - Se extienden hasta un límite calculado con las siguientes fórmulas:  
      - Límite inferior:  
        <div align="center" style="font-size: 24px;">  Q1 - 1.5 x RIQ  </div>  
      - Límite superior:  
        <div align="center" style="font-size: 24px;"> Q3 + 1.5 x RIQ </div>  
    - Los valores fuera de estos límites se consideran **outliers**.

4. **Outliers:**
    - Son los datos que están muy alejados del resto, más allá de los bigotes.
    - Ayudan a identificar posibles errores o valores inusuales en los datos.

### Relación con la Curva de Distribución Normal

En la imagen, se muestra una curva en forma de campana (distribución normal) detrás del boxplot:

- **Zona azul:**  
 Representa los datos que caen dentro de una desviación estándar (𝜎) de la media. Esto incluye aproximadamente el 50% de los datos.

El boxplot está diseñado para reflejar algo similar:

- **Q1 y Q3** en el boxplot dividen el rango en donde se encuentra la mayoría de los datos.
- Los **bigotes** señalan los valores normales.
- Los **puntos fuera de estos bigotes** son considerados **outliers**. (Como lo vimos en el segundo ejercicio donde nos dibuja 3 circulos (1 en la parte inferior y 2 en la parte superior) fuera del boxplot)
