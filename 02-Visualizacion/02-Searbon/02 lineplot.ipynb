```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

## Lineplot

Con lineplot nos permite mostrar información  como tendencias o comportamiento a travez del tiempo.  

Para este ejemplo usaremos 2 listas:  

Para la gráfica usamos la funcion sns.lineplot(x=lista_A, y=lista_B)  

Usamos matplolib para personalizar la gráfica.  

```python
años = [2015, 2016, 2017, 2018, 2019]
total = [45000, 23680, 65984, 78451, 65124]

plt.figure(figsize=(15, 5))
sns.lineplot(x=años, y=total)

plt.title("Evolución de Totales por Año", fontsize=16, fontweight="bold")
plt.xlabel("Años", fontsize=12)
plt.ylabel("Total", fontsize=12)

plt.grid(True)
plt.xticks(años)
```
