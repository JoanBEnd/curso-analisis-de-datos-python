### Preparación para la Limpieza de Datos: Agregar Columnas de Tiempo

Antes de comenzar con la limpieza de los datos, añadiremos tres nuevas columnas al DataFrame:
1. **Fecha**
2. **Año**  
3. **Mes**  
4. **Día**  

Estas columnas serán útiles para realizar análisis basados en el tiempo más adelante.

Para lograr esto, utilizaremos los siguientes comandos:

```python
df["Fecha"] = df["Latest_Launch"].dt.strftime("%Y-%m-%d")
df["Año"] = df["Latest_Launch"].dt.strftime("%Y")
df["Mes"] = df["Latest_Launch"].dt.month_name()
df["Día"] = df["Latest_Launch"].dt.strftime("%d")
```

Expliquemos los elementos utilizados:
1. .dt
    Es una propiedad que nos proporciona pandas para trabajar con columnas de tipo fecha, con el cual nos permite acceder a varias funciones específicas de manejo de datos temporales.

2. strftime()
    Esta función nos ayuda a formatear valores de fecha y hora, permitiendo extraer información específica del campo.  
    Formatos utilizados dentro de la función:  
    
        %Y-%m-%d: Le da un formato de año, mes, dia.  
        %Y: Extrae el año de la fecha.  
        %d: Obtiene el día de la fecha.  

3. month_name()
    Esta función nos retorna el nombre completo del mes en lugar de un número lo cuál es útil para una mejor interpretación visual de los datos.


````python
df_ventas["Fecha"] = df_ventas["Latest_Launch"].dt.strftime("%Y-%m-%d")
df_ventas["Año"] = df_ventas["Latest_Launch"].dt.strftime("%Y")
df_ventas["Mes"] = df_ventas["Latest_Launch"].dt.month_name()
df_ventas["Dia"] = df_ventas["Latest_Launch"].dt.strftime("%d")

# Para no ver todas las columnas, en este caso solo las que hemas creado, las seleccionamos:
df_ventas[['Fecha', 'Año', 'Mes', 'Dia']]
````

 
### Limpieza de Datos: Manejo de Valores Nulos e Índices

#### Eliminación de Valores Nulos:

Para poder limpiar los campos nulos de todas las columnas, podemos usar la siguiente funcion:
```python
df.dropna()
```
Esta función elimina todas las filas que contienen valores nulos en cualquier columna, dejando únicamente las filas con datos completos.  


#### Reseteo de Índices:

Adicional a ello, para poder tener una columna index bien correlacionada, podemos adicionar el reseteo de los indices.  
El código quedaría de la siguiente forma: 
```python
df.dropna().reset_index(drop=True)

````python
#Para poder trabajar mejor, crearemos un nuevo DataFrame, y en ella almacenaremos los datos que no tengan valores nulos
#Evitando asi que modifiquemos el DataFrame original.

df_ventas_limpio = df_ventas.dropna().reset_index(drop=True)
df_ventas_limpio

#A partir de este DataFrame, podemos hacer un análisis de los datos.
````

Para estar más seguros despues de crear el nuevo DataFrame, podemos comprobar si hay valores nulos.

````python
print(df_ventas_limpio.isnull().sum())
````

De la misma forma podemos ver cuantos datos no nulos hay en cada columna, usando la siguiente función:

````python

df_ventas_limpio.notnull().sum()
````

### Analizemos los datos:

Por ejemplo, podemos ver cuantos autos se han lanzado por cada Estado.
1. filtramos por el campo State.
2. contamos los valores.
3. reseteamos los indices.
4. Almacenamos este analisis en un nuevo dataframe

````python
df_state = df_ventas_limpio["State"].value_counts().reset_index()
df_state.columns = ["State", "Total"]
df_state
````


Podemos ver cuantos autos  se han relanzado en cada región
Para ello:
1. filtramos por el campo región
2. contamos los valores
3. reseteamos los indices
4. Almacenamos este analisis en un nuevo dataframe

````python

df_region = df_ventas_limpio["Region"].value_counts().reset_index()
df_region.columns = ["Region", "Total"]
df_region
````

Podemos ver cuantos autos se han lanzando por cada año.
Para ello:
1. filtramos por el campo Año
2. contamos los valores
3. reseteamos los indices
4. Almacenamos este analisis en un nuevo dataframe


````python

df_ventas_lanzamiento_año = df_ventas_limpio["Año"].value_counts().reset_index()
df_ventas_lanzamiento_año.columns = ["Año", "Total"]
df_ventas_lanzamiento_año
````

Podemos hacer un análisis más detallado, por ejemplo, cuantos autos se han lanzado por año y mes.  
Para ello:
1. filtramos por el campo Año y Mes.
2. contamos los valores.
3. reseteamos los indices.
4. Almacenamos este analisis en un nuevo dataframe
5. Modificamos algunos nombres de columnas, y las  que se mantienen, se les agrega el mismo nombre.
6. Podemos ordenar los datos usando las columnas Año y Mes, donde el Año se ordena de manera ascendente y el Total de manera descencente.  
   Eso lo vemos con el parametro ascending = [True, True].


````python

df_ventas_lanzamiento_año_mes = df_ventas_limpio[["Año", "Mes"]].value_counts().reset_index()
df_ventas_lanzamiento_año_mes.columns = ["Año", "Mes", "Total"]
df_ventas_lanzamiento_año_mes.sort_values(by=["Año", "Mes"], ascending=[True, True]).reset_index(drop=True)
````

Podemos ver cuantos autos se han lanzado por año y marca .
Para ello:
1. filtramos por el campo Año y Marca.
2. contamos los valores.
3. reseteamos los indices.
4. Almacenamos este analisis en un nuevo dataframe
5. Modificamos algunos nombres de columnas, y las  que se mantienen, se les agrega el mismo nombre.
6. Podemos ordenar los datos usando las columnas Año y Total, donde el Año se ordena de manera ascendente y el Total de manera descencente.  
   Eso lo vemos con el parametro ascending = [True, False].


````python

df_ventas_marca = df_ventas_limpio[["Año","Manufacturer"]].value_counts().reset_index()
df_ventas_marca.columns = ["Año", "Marca", "Total"]
df_ventas_marca.sort_values(by=["Año", "Total"], ascending=[True, False]).reset_index(drop=True)
````

Podemos saber por marca de carro, que modelo es el más caro.
Para ello:
1. filtramos por el campo Manufacturer, Model y Sales_in_thousands.
2. Agrupamos por Manufacturer.
3. Obtenemos el max().
4. reseteamos los indices.
5. Almacenamos este analisis en un nuevo dataframe
6. Modificamos algunos nombres de columnas, y las  que se mantienen, se les agrega el mismo nombre.
7. Podemos ordenar los datos usando las columnas Precio_Miles, donde el Precio_Miles se ordena de manera descencente.  
   Eso lo vemos con el parametro ascending = False.

````python
df_precio_alto_by_marca = df_ventas_limpio[["Manufacturer","Model","Sales_in_thousands"]].groupby(["Manufacturer"]).max().reset_index()
df_precio_alto_by_marca.columns = ["Marca", "Modelo","Precio_Miles"]
df_precio_alto_by_marca.sort_values(by="Precio_Miles", ascending=False, inplace=True)
df_precio_alto_by_marca.reset_index(drop=True, inplace=True)
df_precio_alto_by_marca
````
