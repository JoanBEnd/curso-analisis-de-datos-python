Que pasa si tengo una columna mal escrita o tiene un nombre que no va con los datos que contiene esa columna, podemos cambiarle el nombre a la columna de la siguiente manera:

Usando la funcion rename(), adentro especificamos la columna a cambiar y la nueva columna y aplicamos el parametro inplace =True  
Ese parametro obliga a que el cambio se haga en el dataframe que ya hemos creado.  
Entonces hacemos lo siguiente:


````python
mi_modalidad.rename(columns={"Unnamed: 0": "Indice"}, inplace=True)
mi_modalidad
````

 
## **Crear una nueva columna basada en una columna existente**

### **Descripción**  
En esta sección, crearemos una nueva columna a partir de una columna existente en el DataFrame.  
Como ejemplo, trabajaremos con la columna **`time`** para extraer los **meses**, pero en lugar de números, utilizaremos los nombres de los meses.

---

### **Pasos Teóricos**  
1. Crear una nueva columna en el DataFrame.  
2. Asignar el proceso necesario para obtener la información desde la columna original.

---

### **Preparación de Fechas con `parse_dates`**  
Para trabajar correctamente con fechas, debemos usar el parámetro **`parse_dates`** al momento de leer el archivo CSV.  
Esto convierte automáticamente las fechas en un formato que **pandas** puede manipular.

### **Ejemplo**  
La lectura del archivo quedaría de la siguiente forma:

```python
pd.read_csv("data/ModalidadVirtual.csv", parse_dates=["time"])
```

¿Por qué usar parse_dates?
- Sin parse_dates, pandas carga las fechas como simples cadenas de texto (str).
- Esto limita la posibilidad de realizar operaciones relacionadas con el tiempo, como:
- Extraer meses o años.
- Realizar cálculos de diferencias de tiempo.
- Filtrar datos basados en fechas.


````python
mi_modalidad["Meses"] = mi_modalidad["time"].dt.month_name()
mi_modalidad
````

Otra forma de crear un nuevo campo es usando condicional, para eso, podemos hacer los siguiente:
Crear una columna Documento donde nos indique que SI los que tienen mayor de 18 y No los menores de 18:

````python
mi_modalidad["Documento"] = [ "Si"  if edad  >= 18 else "No" for edad in mi_modalidad["edad"] ]
mi_modalidad
````

Verifiquemos si los menores de edad tienen asignado la opcion No

````python
mi_modalidad[mi_modalidad["edad"]< 18]
````
